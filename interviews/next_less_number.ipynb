### Задача

Дано положительное целое число target.

Необходимо вернуть также положительное число, наибольшее из возможных, но меньше чем target, которое состоит из тех же цифр (и в том же количестве), что и target. При этом это должно быть 

Дополнительные условия:
1. если не получается вернуть число, которое меньше, то вернуть -1
2. если в итоге первый разряд становится равным 0 - также вернуть -1

Ограничения:

1 <= target <= 2**32-1

Примеры:

- next_less_int(907) --> 790
- next_less_int(1243) --> 1234
- next_less_int(153346) --> 146533
- next_less_int(1234) --> -1
- next_less_int(709) --> -1
- next_less_int(1027) --> -1
- next_less_int(12345679088888) --> 12345678988880
- next_less_int(1234567908) --> 1234567890

```python
def next_less_int(target):
    if target < 10:
        return -1
    result = -1
    # преобразуем в массив для удобства
    target_list = []
    while target:
        target, d = divmod(target, 10)
        target_list.append(d)
    # развернем список 
    l, r = 0, len(target_list)-1
    while l <= r:
        target_list[l], target_list[r] = target_list[r], target_list[l]
        r-=1
        l+=1
    # найдем индекс того разряда, в котором значение больше предыдущего
    for i in range(len(target_list)-2, -1, -1):
        if target_list[i] <= target_list[i+1]:
            continue
        else:
            break  
    pivot_index = i
    x = target_list[pivot_index]
    # найдем максимум нашего хвоста, который меньше x 
    tail_max = -1
    for i in range(len(target_list)-1, -1, -1):
        if target_list[i] < target_list[pivot_index]:
            tail_max = target_list[i]
            break
    if tail_max == -1:
        return result
    target_list[pivot_index], target_list[i] = target_list[i], target_list[pivot_index]
    # sort tail
    if target_list[0] == 0:
        return result
    target_list = target_list[:pivot_index+1]+sorted(target_list[pivot_index+1:], reverse=True)
    # convert list to int
    result = 0
    for i in range(len(target_list)):
        result = result + (10**(len(target_list)-i-1))*target_list[i]
    return result

assert next_less_int(907) == 790
assert next_less_int(1243) == 1234
assert next_less_int(153346) == 146533
assert next_less_int(1234) == -1
assert next_less_int(709) == -1
assert next_less_int(1027) == -1
assert next_less_int(12345679088888) == 12345678988880
assert next_less_int(1234567908) == 1234567890
assert next_less_int(1) == -1
assert next_less_int(10) == -1
```
